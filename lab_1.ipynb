```python
def predict_sentence(start_word, length=15):
    # Ensure start word is in the format our Counters expect (string)
    current_word = start_word
    sentence = [current_word]
    
    # Vocabulary size for smoothing (optional in generation, but good for context)
    vocab_size = len(unigram_counts)

    print(f"Generating sentence starting with: '{start_word}'...")

    for _ in range(length - 1):
        # --- Step A: Find potential next words based on Bigrams ---
        # Look for keys in bigram_counts that look like: (current_word, next_word)
        candidates = [key[1] for key in bigram_counts.keys() if key[0] == current_word]
        
        if candidates:
            # If we found bigrams, pick one based on their frequency (probability)
            candidate_weights = [bigram_counts[(current_word, w)] for w in candidates]
            
            # random.choices picks a word using the counts as probability weights
            next_word = random.choices(candidates, weights=candidate_weights, k=1)[0]
        else:
            # --- Step B: Backoff to Unigrams ---
            # If the current word has no history (it's new or end of sentence), 
            # pick a random word from the unigrams.
            # Note: unigram_counts keys are tuples like ('word',), so we take k[0]
            all_words = [k[0] for k in unigram_counts.keys()]
            next_word = random.choice(all_words)
        
        # Append and update for next loop
        sentence.append(next_word)
        current_word = next_word

    return " ".join(sentence)

```
